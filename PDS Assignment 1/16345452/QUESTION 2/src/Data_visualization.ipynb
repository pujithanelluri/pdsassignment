```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Create a DataFrame from the given data
# Specify the file path to your CSV file
modified_studentPerformance_data = '/content/drive/Othercomputers/PDS Assignment 1/16345452/QUESTION 2/Stage 2(Cleaned_Data)/cleaned_data.csv'
# Use pandas to read the CSV file into a DataFrame
df = pd.read_csv(modified_studentPerformance_data)

#output_txt_file = '/content/drive/Othercomputers/PDS Assignment 1/16345452/QUESTION 2/Stage 3 (Results)/results.txt'
#resultFile = open(output_txt_file, 'w')

#o_f = '/content/drive/Othercomputers/PDS Assignment 1/16345452/QUESTION 2/Stage 3 (Results)'


# Calculate the average scores for each subject by gender
gender_scores = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean()

# Plot the bar chart
gender_scores.plot(kind='bar', figsize=(10, 6))
plt.title('Student Performance by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.legend(title='Subject', loc='upper right')
plt.show()
```

```python
# Group by race/ethnicity and calculate the average scores
race_scores = df.groupby('race/ethnicity')[['math score', 'reading score', 'writing score']].mean()

# Transpose the DataFrame for stacking
race_scores = race_scores.T

# Plot the stacked bar chart
race_scores.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.title('Student Performance by Race/Ethnicity')
plt.xlabel('Subject')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.legend(title='Race/Ethnicity', loc='upper right')
plt.show()
```

```python
# Create a violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='parental level of education', y='math score', data=df, inner='quart', cut=0)
plt.title('Math Score Distribution by Parental Education Level')
plt.xlabel('Parental Education Level')
plt.ylabel('Math Score')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x='test preparation course', y='math score', data=df)
plt.title('Math Score Distribution by Test Preparation Course')
plt.xlabel('Test Preparation Course')
plt.ylabel('Math Score')
plt.show()




```

```python
# Calculate the correlation matrix
correlation_matrix = df[['math score', 'reading score', 'writing score']].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Marks (Math, Reading, Writing)')
plt.show()
```
