#B)Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from “New_price”).

```python
df1=pd.DataFrame(df)
```

```python
df1['Mileage'] = df1['Mileage'].str.extract('(\d+\.\d+)').astype(float)

df1['Engine'] = df1['Engine'].str.replace(' CC', '').astype(float)

df1['Power'] = df1['Power'].str.extract('(\d+\.\d+)').astype(float)

df1['New_Price'] = df1['New_Price'].str.extract('(\d+\.\d+)').astype(float)
```

# A)Look for the missing values in all the columns and either impute them (replace with mean, median, or mode) or drop them. Justify your action for this task

```python
#checking for the missing values
missing_values = df.isnull()
print(missing_values)
```

```python
#imputing the missing values
missing_values = df.isnull()
print(missing_values)
```

```python
df_filled_mean = df.fillna(df.mean())
print(df_filled_mean)
```

```python
#imputing from median
df_filled_median = df.fillna(df.median())
print(df_filled_median)
```

```python
#imputing with mode
df_filled_mode = df.fillna(df.mode().iloc[0])
print(df_filled_mode)
```

```python
#dropping the missing values
df_dropped = df.dropna()
print(df_dropped)
```

```python
df1.isna().sum()
```

**RESULTS:**

Imputation with mean, median, or mode is used to preserve the overall
1. statistical properties of the data and can be especially helpful when the missing data is missing at random. It's a simple and effective way to handle missing values without losing a significant amount of information.
2. Dropping missing values is justifiable when the missing data is not informative and the loss of data does not significantly impact the analysis. It should be avoided if missing data patterns carry important information.

#C)Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value

```python
#To know the categorical values of fuel type and transmission
df['Fuel_Type'].unique()
#setting Diesel =1,Petrol=2, Electric=3
```

```python
#we can set manual as '0' and automatic as '1'
df['Transmission'].unique()
```

```python
df1.head()
```

```python
#Relacing Categorical variables into numerical one hot encoded value
df['Fuel_Type'].replace({'Diesel': 0, 'Petrol': 1, 'Electric': 2}, inplace=True)
df['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
```
